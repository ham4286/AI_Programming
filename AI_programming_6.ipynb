```python
import requests
import IPython.display as display
import base64

url = "https://stable-diffusion-demo-kk0powt97tmb.octoai.cloud/text2img"

headers = {
    "content-type": "application/json"
}

data = {
    "prompt": "battleground",
    "negative_prompt": "Blurry photo, distortion, low-res, bad quality",
    "guidance_scale": 7.5,
    "strength": 0.8,
    "num_inference_steps": 30,
    "model": "DreamShaper",
    "lora": {"LowRA": 0.5},
    "text_inversion": {"ng_deepnegative_v1_75t": "ng_deep"},
    "height": 512,
    "width": 512,
    "scheduler": "DPMSolverMultistep"
}

```

```python
response = requests.post(url, headers=headers, json=data)

if response.status_code == 200:
    image_data = response.json().get("completion", {}).get("image_0", "")
    if image_data:
        image_bytes = base64.b64decode(image_data)
        display.display(display.Image(data=image_bytes))
    else:
        print("Image data not found in the response.")
else:
    print(f"Request failed with status code: {response.status_code}")
    print(response.text)

```
